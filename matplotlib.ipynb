```python
from  matplotlib import pyplot as plt
```

```python
fig=plt.figure()
axis=fig.add_axes([0.5,0.1,0.8,0.8]) # control the left ,right ,width,height of the canvas(from 0 to 1)
#fig.add_axes() initialises a subplot
axis.plot(x,y,'b--')
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2) #submap is of 1 row 2 columns
for ax in axes:
    ax.plot(x,y,'r')
```

```python
# we can plot a graph inside another graph
fig=plt.figure()
# control the left,r,w,h of the canvas
axes1=fig.add_axes([0.1,0.1,0.8,0.8]) #big axes
axes2=fig.add_axes([0.2,0.5,0.4,0.3])#small canvas
axes1.plot(x,y,'g')
axes2.plot(y,x,'y')
```

```python
fig=plt.figure(figsize=(16,9),dpi=300) #enlarged graph to print accordingly
axis=fig.add_axes([0.5,0.1,0.8,0.8]) 
axis.plot(x,y,'g')
```

```python
fig,axes=plt.subplots()
axes.set_title("SIMPLE GRAPH")
axes.set_xlabel("X")
axes.set_ylabel("Y")
axes.plot(x,y,'r')
```

```python
fig,ax=plt.subplots(dpi=150)

ax.plot(x,x+1,color="blue" , lw=0.25)
ax.plot(x,x+2,color="blue" , linewidth=1)
ax.plot(x,x+3,color="blue" , linewidth=1.5)
ax.plot(x,x+4,color="blue" , lw=2,linestyle='-')
ax.plot(x,x+5,color="blue" , lw=2,linestyle=':')
line,=ax.plot(x,x+6,color="black" , lw=1.50,linestyle='-.')
ax.plot(x,x+8,color="blue",marker="o",markersize=8,markerfacecolor="red")
line.set_dashes([5,10,15,20])
ax.plot(x,x+7,color="green",marker="s")
ax.plot(x,x+9,color="orange",marker=">")
```

```python
n=np.array([0,1,2,3,4,5])
fig,axes=plt.subplots(1,4,figsize=(16,5))
axes[0].set_title("scatter")
axes[0].scatter(x,x**2)

axes[1].set_title("step")
axes[1].step(n,n**2)

axes[2].set_title("bar")
axes[2].bar(n,n**2,align="center",width=0.5,alpha=0.5)

axes[3].set_title("Fill between")
axes[3].fill_between(x,x**2,x**3,color="green",alpha=0.5)
```

```python
#draw radar chart



fig=plt.figure(figsize=(6,6))
ax=fig.add_axes([0.0,0.0,1,1],polar=True)
t=np.linspace(0,2*np.pi,100)
ax.plot(t,t,color="blue",lw=3)
```

```python
# histogram

n=np.random.randn(100000)
fig,axes=plt.subplots(1,2,figsize=(12,4))
axes[0].set_title("Default histogram")
axes[0].hist(n)

axes[1].set_title("cumulative detailed histogram")
axes[1].hist(n,cumulative=True,bins=5)
```
